                                    STACK DATASTRUCTUR

```python
class Stack():
    def __init__(self):
        self.stack=[]

    def push(self,data):
        '''
        Function takes the data and and push to the stack
    
        parameters data-user input 

        return -None
       '''
        self.stack.append(data)

    def pop(self):
        '''Function takes the data and and push to the stack
    
        parameters -None

        return -None
        '''
        self.stack.pop()
    
    def length(self):
        '''Function find the length of the stack
    
        parameters -None

        return int-return the length of the stack

        '''
        return len(self.stack)
    
    def show(self):
        '''Function shows the all elements.
    
        parameters -None

        return -None
        '''
        print(self.stack)

    
    def is_Empty(self):
        '''Function check the stack is empty or not
    
        parameters -None

        return -None
        '''
        if len(self.stack)==0:
            return True
        return False

stack=Stack()


stack.push(9)
stack.push(10)
stack.push(1)

stack.pop()
stack.pop()
stack.pop()
stack.show()
print(stack.is_Empty())
```

### 1.program to check the correct paranthasis or not

```python
def is_valid(s):
    '''
    Function take the string input and check the is it close or not
    
    parameters str-user input 

    return -boolean-it return the true or false value
    '''
    bracket_map = {')': '(', '}': '{', ']': '['}
    stack = []

    for char in s:
        if char in bracket_map:
            top_element = stack.pop() if stack else '#'
            if bracket_map[char] != top_element:
                return False
        else:
            stack.append(char)

    return not stack

def main():
    try:
        s=input("Enter check the brackets")
        if is_valid(s):
            print("the given paranthasis is valid")
        else:
            print("The given paranthasis is invalid")
    except ValueError as e:
        print('Enter correct value',e)

if __name__ == '__main__':
    main()
```

### 2.program to reverse a string using the stack

```python
def rev_string(string):
    '''
    Function take the string input and reverse string and return it
    
    parameters str-user input 

    return -boolean-it return the true or false value
    '''
    stack = []

    for char in string:
        stack.append(char)

    rev=" "
    while stack:
         rev += stack.pop()
        
    return rev

def main():
    s='sanjay'

    print("Reverse of string is:",rev_string(s))


if __name__ == '__main__':
    main()
```

### 1.program to print the binary values to upto given number.

```python
from collections import deque
def generatePrintBinary(n):
    '''
    Function takes the a number a prints binary values

    parameters int-number

    return -None
    '''

    q = deque()
    
    q.append("1")
    
    for _ in range(n):
        binary_number = q.popleft()
        
        print(binary_number)
        
        q.append(binary_number + "0")
        q.append(binary_number + "1")


def main():
    n = 10
    generatePrintBinary(n)


if __name__ == "__main__":
    main()
```

```python
class PrinterQueue:
    def __init__(self):
        self.queue = []

    def add_print_job(self, job):
        '''function asign to the next job to printer queue
         parameters-job
         return-None
        '''
        self.queue.append(job)
        print(f"Print job '{job}' added to the queue.")

    def process_print_job(self):
        '''function check the printer is free or not
        parameters-None
        return-None
        '''
        if self.is_empty():
            print("No print jobs to process, the queue is empty.")
            return None
        job = self.queue.pop(0)
        print(f"Print job '{job}' is being processed.")
        return job

    def next_print_job(self):
        if self.is_empty():
            print("The queue is empty.")
            return None
        return self.queue[0]

    def is_empty(self):
        '''function check the printer is free or not
        parameters-None
        return-None
        '''
        return len(self.queue) == 0
    

def main():
    printer_queue = PrinterQueue()
    printer_queue.add_print_job("Document1")
    printer_queue.add_print_job("Document2")
    print(printer_queue.next_print_job())  
    printer_queue.process_print_job()     
    print(printer_queue.next_print_job())  
    print(printer_queue.is_empty())       
    printer_queue.process_print_job()      
    print(printer_queue.is_empty())    


if __name__ == "__main__":
    main()   
```

                                              Tree datastructures

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
        self.parent = None

    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent

        return level

    def print_tree(self):
        spaces = ' ' * self.get_level() * 3
        prefix = spaces + "|__" if self.parent else ""
        print(prefix + self.data)
        if self.children:
            for child in self.children:
                child.print_tree()

    def add_child(self, child):
        child.parent = self
        self.children.append(child)

def build_product_tree():
    root = TreeNode("Electronics")

    laptop = TreeNode("Laptop")
    laptop.add_child(TreeNode("Mac"))
    laptop.add_child(TreeNode("Surface"))
    laptop.add_child(TreeNode("Thinkpad"))

    cellphone = TreeNode("Cell Phone")
    cellphone.add_child(TreeNode("iPhone"))
    cellphone.add_child(TreeNode("Google Pixel"))
    cellphone.add_child(TreeNode("Vivo"))

    tv = TreeNode("TV")
    tv.add_child(TreeNode("Samsung"))
    tv.add_child(TreeNode("LG"))

    root.add_child(laptop)
    root.add_child(cellphone)
    root.add_child(tv)

    root.print_tree()

if __name__ == '__main__':
    build_product_tree()
```

                                                   Binary Tree

```python
class BinarySearchTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def add_child(self, data):
        if data == self.data:
            return # node already exist

        if data < self.data:
            if self.left:
                self.left.add_child(data)
            else:
                self.left = BinarySearchTreeNode(data)
        else:
            if self.right:
                self.right.add_child(data)
            else:
                self.right = BinarySearchTreeNode(data)


    def search(self, val):
        if self.data == val:
            return True

        if val < self.data:
            if self.left:
                return self.left.search(val)
            else:
                return False

        if val > self.data:
            if self.right:
                return self.right.search(val)
            else:
                return False

    def in_order_traversal(self):
        elements = []
        if self.left:
            elements += self.left.in_order_traversal()

        elements.append(self.data)

        if self.right:
            elements += self.right.in_order_traversal()

        return elements


def build_tree(elements):
    print("Building tree with these elements:",elements)
    root = BinarySearchTreeNode(elements[0])

    for i in range(1,len(elements)):
        root.add_child(elements[i])

    return root

if __name__ == '__main__':
    countries = ["India","Pakistan","Germany", "USA","China","India","UK","USA"]
    country_tree = build_tree(countries)

    print("UK is in the list? ", country_tree.search("UK"))
    print("Sweden is in the list? ", country_tree.search("Sweden"))

    numbers_tree = build_tree([17, 4, 1, 20, 9, 23, 18, 34])
    print("In order traversal gives this sorted list:",numbers_tree.in_order_traversal())
```

```python
def bfs(graph, node):
    '''Function takes the graph and node and retruns all values
    
    parameters dict,str-graph,str
    
    return -None'''
    visited = []
    queue = []
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)
        print(m, end=" ")

        for neighbor in graph[m]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)

def main():
    graph = {
        5: [2, 4, 6, 3],
        3: [3, 5],
        1: [9, 0],
        2: [2],
        4: [],  
        6: [], 
        9: [],  
        0: []   
    }
    print("The following is Breadth-First Search")
    bfs(graph, 5)

if __name__ == '__main__':
    main()
```

```python
def dfs(graph, visited, root):
    '''Function takes the graph and node and retruns all values
    
    parameters dict,str-graph,str
    
    return -None
    '''
    if root not in visited:
        print(root)
        visited.add(root)
        for neighbor in graph[root]:
            dfs(graph, visited, neighbor)

def main():
    graph = {
        'a': ['b', 'c', 'f'],
        'b': ['c', 'm'],
        'c': [],
        'f': [],
        'm': []
    }
    visited = set()
    print("The following is Depth-First Search")
    dfs(graph, visited, 'a')

if __name__ == '__main__':
    main()
```
